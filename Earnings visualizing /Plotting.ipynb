## Q and A

> **Do students in more popular majors make more money?**

```python
recent_grads.plot(x="Total", y="Median", kind="scatter")
```

*No, students in more popular majors take mid-paid.*

> **Do students that majored in subjects that were majority female make more money?**

```python
recent_grads[recent_grads["Women"] > recent_grads["Men"]].plot(y="Median", kind="hist")
```

*From the previous case, we can see that the median holds in the range of 20,000 to 80,000. In this case we see, that the main part is slightly less than 40 000. *

> **What percent of majors are predominantly male? Predominantly female?**

```python
cut_rg= pd.cut(recent_grads["ShareWomen"], bins=[0, 0.5, 1], include_lowest=True)
cut_norm_rg = cut_rg.value_counts(sort=False, normalize=True).mul(100)
ax = cut_norm_rg.plot.bar(rot=0, color="b", figsize=(10, 5))

ax.set_xticklabels(["Male", "Female"])

plt.title("Relation of majors predominantly male and predominantly female")
plt.ylabel("%, percentages")
```

**Note**: we believe that "predominantly" is more than half.

Male: ±44%;
Female: ±56.

> **What's the most common median salary range?**

```python
pop_salary = recent_grads["Median"].value_counts().index.tolist()[0]
next_salary = recent_grads["Median"][recent_grads["Median"] > pop_salary].min()
prev_salary = recent_grads["Median"][recent_grads["Median"] < pop_salary].max()

ax_range = []
ax_range.append(prev_salary)
ax_range.append(next_salary)

ax = recent_grads["Median"].hist(bins=3, figsize=(5,5), range=ax_range)
```

The most popular median is range [34000, 35600] or [prev_salary, next_salary].
